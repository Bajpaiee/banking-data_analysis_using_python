```python
#examine the distrubution of unique cataegorical colums
categorical_cols = df[["BRId", "IAId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns
for col in categorical_cols:
    print(f"Value counts for '{col}':")
    display(df[col].value_counts())
```

Univariates Analysis

```python
for i, predictor in enumerate(df[["BRId", "IAId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor, hue='GenderId')
```

#Bivariate Analysis

```python
for i, predictor in enumerate(df[["BRId", "IAId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor, hue='Nationality')
```

```python
#HIplot of value counts for different occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(10,4))
    sns.histplot(df[col])
    plt.title('Histrogram of occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

#Numerical Analysis

```python
Numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

#univarient analysis and visualization
plt.figure(figsize = (15,10))
for i,col in enumerate(Numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)

plt.tight_layout()  # optional: avoids overlapping
plt.show()

```

#Heatmap

```python
Numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
 
correction_matrix= df[Numerical_cols].corr()
plt.figure(figsize = (12,12))
sns.heatmap(correction_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title("Correction Matrix")
plt.show()

```

#Insight of EDA
The strongest positive  correction occur among 'bank deposits'with 'checking accounts','Saving accounts',indicating that customers who maintain high  balance in one account indicating  that customer who  maintain high balance in one account type often hold substantail amount / fund across other account as well
